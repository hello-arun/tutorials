```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

def init_figure(shape=(1,1),size=[89,89]):
    fig,axes = plt.subplots(nrows = shape[0], ncols = shape[1],squeeze=False)
    fig.set_size_inches(size[0]/25.4,size[1]/25.4)
    return (fig,axes)
```

```python
def exp_growth(t,y,k=1.0):
    return k*y

solution = solve_ivp(
    fun = exp_growth,
    t_span = [0,10],  # t0,t1
    y0 = [1],  # y0 is y at t0
    method = "RK45",
    t_eval=np.linspace(0,10,40)
    )

fig,[[ax]]=init_figure()
ax.plot(solution.t,solution.y[0])
```

```python
def log_growth(t,y,r=1.0,k=100.0):
    # y: population size
    # r: growth rate
    # k: carrying capacity
    return r*y*(1.0-y/k)

solution = solve_ivp(
    fun = log_growth,
    t_span = [0,10],  # t0,t1
    y0 = [1],  # y0 is y at t0
    method = "RK45",
    t_eval=np.linspace(0,10,40)
    )

fig,[[ax]]=init_figure()
ax.plot(solution.t,solution.y[0])
```
